# Modifying Image Pixels

```python
cb_img_copy=cb_image.copy()
cb_img_copy[2,2]=200
cb_img_copy[2,3]=200
cb_img_copy[3,2]=200
cb_img_copy[3,3]=200

plt.imshow(cb_img_copy,cmap='gray')
print(cb_img_copy)
```

```python
cb_img_copy=cb_image.copy()
cb_img_copy[2:4,2:4]=200

plt.imshow(cb_img_copy,cmap='gray')
print(cb_img_copy)
```

```python
img_NZ=cv2.imread('New_Zealand_Boat.jpg',cv2.IMREAD_COLOR)
```

```python
img_NZ=img_NZ[:,:,::-1]
plt.imshow(img_NZ)
```

```python
img_NZ.shape
```

```python
cropped_region=img_NZ[200:400,300:600]
plt.imshow(cropped_region)
```

```python
cropped_region=cropped_region[40:120,110:160]
plt.imshow(cropped_region)
```

```python
resize_cropped_2x=cv2.resize(cropped_region,None,fx=2,fy=2)
plt.imshow(resize_cropped_2x)
```

```python
desired_width=100
desired_height=200
dim=(desired_width,desired_height)

resize_cropped_desired=cv2.resize(cropped_region,dsize=dim,interpolation=cv2.INTER_AREA)
plt.imshow(resize_cropped_desired)
```

```python
desired_width=100
espect_ratio=desired_width/cropped_region.shape[1]
desired_height=int(cropped_region.shape[0]*espect_ratio)
dim=(desired_width,desired_height)

resize_cropped_ratio=cv2.resize(cropped_region,dsize=dim,interpolation=cv2.INTER_AREA)
plt.imshow(resize_cropped_ratio)
```

```python
resize_cropped_ratio=resize_cropped_ratio[:,:,::-1]
cv2.imwrite("Test.png",resize_cropped_ratio)
Image(filename="Test.png")
```

```python
img_NZ_flipped_h=cv2.flip(img_NZ,1)
img_NZ_flipped_v=cv2.flip(img_NZ,0)
img_NZ_flipped_both=cv2.flip(img_NZ,-1)

plt.figure(figsize=[18,5])

plt.subplot(141);plt.imshow(img_NZ_flipped_h);plt.title("Horizontal Flip");
plt.subplot(142);plt.imshow(img_NZ_flipped_v);plt.title("Vertical Flip");
plt.subplot(143);plt.imshow(img_NZ_flipped_both);plt.title("Both Flipped");
plt.subplot(144);plt.imshow(img_NZ);plt.title("Original");
```
